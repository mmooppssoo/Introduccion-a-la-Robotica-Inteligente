```python
import matplotlib.pyplot as plt
import numpy as np

# Representación de la arena
arena = [
    "%%%%%%%%%%%%%%%%%%%%",
"%###%##############%",
"%###%##############%",
"%###%##########%%%%%",
"%##################%",
"%##################%",
"%##%%%%%####%%%%%##%",
"%######%###########%",
"%######%###########%",
"%%%%%##%####%%%%###%",
"%######%####%######%",
"%######%####%######%",
"%##%%%%%####%######%",
"%###########%%%%###%",
"%##################%",
"%####%#############%",
"%####%########%%%%%%",
"%####%#############%",
"%####%#############%",
"%%%%%%%%%%%%%%%%%%%%",

]

# Convertimos la representación en una matriz de valores
arena_matrix = np.array([[1 if cell == "%" else 0 for cell in row] for row in arena])

# Creando la figura
fig, ax = plt.subplots(figsize=(8, 8))

# Dibuja la arena (muros y huecos)
for i in range(len(arena)):
    for j in range(len(arena[i])):
        if arena_matrix[i, j] == 1:  # Muro
            ax.add_patch(plt.Rectangle((j, len(arena) - i - 1), 1, 1, color='lightgrey'))
        else:  # Espacio libre
            ax.add_patch(plt.Rectangle((j, len(arena) - i - 1), 1, 1, color='white'))

# Colocamos los puntos rojos en el centro de cada celda libre
for i in range(len(arena)):
    for j in range(len(arena[i])):
        if arena_matrix[i, j] == 0:  # Espacio libre
            ax.plot(j + 0.5, len(arena) - i - 0.5, 'ro', markersize=5)

# Configuración de los ejes con los valores dentro de las cuadriculas
# Valores para el eje y (invertido)
y_ticks = np.linspace(1.425, -1.425, len(arena))
# Valores para el eje x (invertido y corregido para que vaya de negativo a positivo de izquierda a derecha)
x_ticks = np.linspace(-1.425, 1.425, len(arena[0]))

# Ajustamos los valores de los ejes
ax.set_xticks(np.arange(0.5, len(arena[0]), 1))  # Se coloca un tick en el centro de cada cuadrícula
ax.set_yticks(np.arange(0.5, len(arena), 1))  # Lo mismo para el eje y

# Añadir los valores de los índices en los ejes
ax.set_xticklabels([f"{x:.3f}" for x in x_ticks], fontsize=6)
ax.set_yticklabels([f"{y:.3f}" for y in y_ticks], fontsize=6)

# Desactivar los "minor ticks" (pequeñas marcas) entre los valores numéricos
ax.tick_params(axis='x', which='minor', size=0)
ax.tick_params(axis='y', which='minor', size=0)

# Limitar los ejes
ax.set_xlim(0, len(arena[0]))
ax.set_ylim(0, len(arena))

# Invertir el eje Y para reflejar la imagen en espejo vertical
ax.invert_yaxis()

# Aspecto cuadrado
ax.set_aspect('equal')

# Mostrar el gráfico
plt.show()
```
