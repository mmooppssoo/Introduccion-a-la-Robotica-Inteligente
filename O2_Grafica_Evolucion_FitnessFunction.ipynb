```python

# # Evolución de la Función de Fitness
# Esta celda carga los datos de fitness.log y dibuja las curvas de fitness Mejor, Media y Peor.

import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos: columnas sin encabezado, separadas por espacios/tabs
# Las columnas son: Generación, Mejor, Media, Peor
df = pd.read_csv('fitness.log',
                 sep=r'\s+',
                 header=None,
                 comment='#',
                 names=['Generación', 'Mejor', 'Media', 'Peor'])

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar cada serie con su estilo
plt.plot(df['Generación'], df['Mejor'],   linestyle='-',  label='Best')
plt.plot(df['Generación'], df['Media'],   linestyle='--', label='Average')
plt.plot(df['Generación'], df['Peor'],    linestyle=':',  label='Worst')

# Etiquetas y leyenda
plt.xlabel('Generations', fontsize=12)
plt.ylabel('Fitness',    fontsize=12)
plt.title('Evolución de la Fitness por Generación', fontsize=14)
plt.legend(loc='lower right', frameon=True)

# Ajustes estéticos
plt.xlim(df['Generación'].min(), df['Generación'].max())
plt.ylim(0, 1.0)  # opcional, si quieres forzar 0–1 en y

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
