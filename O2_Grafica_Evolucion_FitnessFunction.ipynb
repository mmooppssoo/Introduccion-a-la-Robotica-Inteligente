```python

# # Evolucin de la Funcin de Fitness
# Esta celda carga los datos de fitness.log y dibuja las curvas de fitness Mejor, Media y Peor.

import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos: columnas sin encabezado, separadas por espacios/tabs
# Las columnas son: Generacin, Mejor, Media, Peor
df = pd.read_csv('fitness.log',
                 sep=r'\s+',
                 header=None,
                 comment='#',
                 names=['Generacin', 'Mejor', 'Media', 'Peor'])

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar cada serie con su estilo
plt.plot(df['Generacin'], df['Mejor'],   linestyle='-',  label='Best')
plt.plot(df['Generacin'], df['Media'],   linestyle='--', label='Average')
plt.plot(df['Generacin'], df['Peor'],    linestyle=':',  label='Worst')

# Etiquetas y leyenda
plt.xlabel('Generations', fontsize=12)
plt.ylabel('Fitness',    fontsize=12)
plt.title('Evolucin de la Fitness por Generacin', fontsize=14)
plt.legend(loc='lower right', frameon=True)

# Ajustes estticos
plt.xlim(df['Generacin'].min(), df['Generacin'].max())
plt.ylim(0, 1.0)  # opcional, si quieres forzar 01 en y

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
